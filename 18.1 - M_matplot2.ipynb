## Bar plot: Bivariate
    - It is used to compare one class with other
    - ex. if we want to compare covid19 cases statewise

```python
state = ['MAH','KA','GOA']
cases = [90,45,30]
```

```python
plt.bar(state,cases)
# default width is: .8
# default color is Blue
# default alignment is : center
```

```python
plt.bar(state,cases,width=0.4,align='edge')
# we have changed alignment to edge and reduced width by 50 %
```

```python
plt.bar(state,cases,width=0.4,color=['r','y','c'])
```

```python
state
```

```python
cases
```

```python
cases22 = [56,80,34]
cases22
```

```python
# lets compare 2019 and 2022 cases using bar plot
plt.bar(state,cases,color='r',label='2019')
plt.bar(state,cases22,color='y',label='2022')
plt.legend()
```

```python
cases,cases22
```

```python
plt.bar(state,cases,color='r',label='2019',width=0.4,align='edge')
plt.bar(state,cases22,color='y',label='2022',width=0.4)
plt.legend()
```

```python
# I want clear separation of 2019 and 2022 data
import numpy as np
import matplotlib.pyplot as plt
data1 = [10,20,30]
data2 = [20,30,40]
names = ['a','b','c']
names_axis = np.arange(len(names))
plt.bar(names_axis - 0.2,data1,color ='red',width = 0.4,label = 'DATA1')
plt.bar(names_axis + 0.2,data2,color = 'pink',width = 0.4,label = 'DATA2')
plt.legend()
plt.xlabel('NAMES')
plt.ylabel('DATA')
plt.title('NAMES VS DATA')
plt.xticks(ticks=range(3),labels = ['a','b','c'])
plt.show()
```

## barh(): Horizontal bar plot

```python
#Horizontal bar plot
# the change is: instead of width we have to use height 
plt.barh(state,cases,color='r',label='2019',height=0.4,align='edge')
plt.barh(state,cases22,color='y',label='2022',height=0.4)
plt.legend()
```

```python
state[::-1]
```

```python
plt.barh(state[::-1],cases[::-1],color='r',label='2019',height=0.4,align='edge')
plt.barh(state[::-1],cases22[::-1],color='y',label='2020',height=0.4)
plt.legend()
```
