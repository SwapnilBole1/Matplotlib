## % distribution of data

## univariate plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ABD data
runs = ['1s','2s','3s','4s','6s']
data = [12,4,8,9,6]
```

```python
plt.pie(data,labels = runs)
plt.show()
```

```python
plt.pie(data, labels = ('singles','doubles','3runs','fours','sixes'))
plt.show()
```

```python
# show the % distribution 
plt.pie(data,labels =runs, autopct='%f%%')
plt.show()
```

```python
# show the % distribution 
plt.pie(data,labels =runs, autopct='%.2f%%')
plt.show()
```

```python
# limit the %
plt.pie(data,labels = runs, autopct='%.f%%')
plt.show()
```

```python
# change color
plt.pie(data, labels = runs, autopct='%.f%%', colors=['r','y','g','c','m'])
plt.show()
```

```python
# change label distance
plt.pie(data,labels = runs,autopct='%.f%%',labeldistance=0.4)
plt.show()
```

```python
plt.pie(data, labels=runs, autopct='%.f%%', labeldistance=0.6, pctdistance=1.1)
plt.show()
```

```python
# want to separate pie
plt.pie(data, labels = runs, autopct='%.f%%',labeldistance=0.4,pctdistance=1.1,explode=[0,0,0,0,0.2], radius=1.5)
plt.show()
```

```python
# want to separate pie
plt.pie(data, labels = runs, autopct='%.f%%',labeldistance=0.4,pctdistance=1.1,explode=[0,0,0.2,0,0], radius=1.5)
plt.show()
```

### When we want to show multiple plot in a single canvas as a separate plot then use subplots

# Subplots

```python
# 1 row and 2 columns
plt.subplots(1,2)
```

```python
# 2 rows and 1 column
plt.subplots(2,1)
```

```python
# 2 rows and 2 Columns
plt.subplots(2,2)
```

```python
fig, ax = plt.subplots(1,2)
fig,ax
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1 = plt.bar(runs, data)
ax2 = plt.plot(runs, data)
```

```python
# create two subplots and unpack the output array immediately
f, (ax1,ax2) = plt.subplots(1,2, sharey = True)

ax1.plot(runs,data)
ax1.set_title('plot1')

ax2.scatter(runs,data)
ax2.set_title('plot2')

plt.suptitle('graph')
```

```python
f, (ax1,ax2) = plt.subplots(1,2)

ax1.plot(runs,data)
ax1.set_title('plot1')

ax2.scatter(runs,data)
ax2.set_title('plot2')

plt.suptitle('graph')
plt.tight_layout()
```

```python
# create four polar axes and access them through the returned array
fig,axs = plt.subplots(2,2)
axs[0,1].plot(runs, data)
axs[1,0].scatter(runs,data)
plt.tight_layout()
```

# Simplest Way to subplot

```python
plt.subplot(121)
plt.plot(runs,data)

plt.subplot(122)
plt.plot(runs,data, 'r--')
```

```python
plt.subplot(221)
plt.plot(runs,data)

plt.subplot(224)
plt.bar(runs,data)
```

```python
plt.figure(figsize=(10,8))
plt.subplot(221)
plt.plot(runs,data)

plt.subplot(224)
plt.bar(runs,data)
```
