```python
#Uni/Bivariate
score = [12,23,4,5,12,30]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt
```

```python
#for line plot just use plot() function
print(score)
plt.plot(score)

#score will b on y axis
# index will be on X axis
```

```python
score
```

```python
# set limit for X and Y axis
plt.plot(score)
plt.xlim((0,8))
plt.ylim((1,24))
```

```python
score2 = [4,10,14,15,22,28]
```

```python
plt.plot(score2)
```

```python
#combine plot 1 and plot 2
plt.plot(score)
plt.plot(score2)
# to change label of X axis from int to object
plt.xticks(ticks=range(6),labels=['Day1','Day2','Day3','Day4','Day5','Day6'],rotation=20)
plt.show()
```

```python
# Use indicators to display info of each graph
plt.plot(score,label='Maharashtra')
plt.plot(score2,label='Andhra')
plt.xticks(ticks=range(6),labels=['Day1','Day2','Day3','Day4','Day5','Day6'])
plt.legend(loc = 'lower right')
plt.show()
```

```python
# Provide x & y axis info and title too
plt.plot(score,label='Maharashtra')
plt.plot(score2,label='Andhra')
plt.xticks(ticks=range(6),labels=['Day1','Day2','Day3','Day4','Day5','Day6'])

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Day vs Temp')

plt.legend()
plt.show()
```

```python
#change color= blue(default),orange,yellow,green,red,black,magenta,cyan
plt.plot(score,label='Maharashtra',color='#57b7e7')
plt.plot(score2,label='Andhra',color='#f348f3')
plt.xticks(ticks=range(6),labels=['Day1','Day2','Day3','Day4','Day5','Day6'])
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Day vs Temp')
plt.legend()
plt.show()
```

```python
plt.plot(score,label='Maharashtra',color='#57b7e7')
plt.plot(score2,label='Andhra',color='#f348f3')
plt.xticks(ticks=range(6),labels=['Day1','Day2','Day3','Day4','Day5','Day6'])
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Day vs Temp')
plt.legend()
plt.savefig('sample.jpg')#this wont give proper resolution of an img
plt.show()
```

```python
plt.plot(score,label='Maharashtra',color='#57b7e7')
plt.plot(score2,label='Andhra',color='#f348f3')
plt.xticks(ticks=range(6),labels=['Day1','Day2','Day3','Day4','Day5','Day6'])
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Day vs Temp')
plt.legend()

# in order to get proper resolution use dpi(dots per inch)
plt.savefig('sample.jpg',dpi=300)
plt.show()
```

```python
plt.plot(score,label='Maharashtra',color='#761377',linestyle='--',linewidth=4)# dashed lines
plt.plot(score2,label='Andhra',color='#146f64',linestyle='--',linewidth=4)
plt.xticks(ticks=range(6),labels=['Day1','Day2','Day3','Day4','Day5','Day6'])
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Day vs Temp')
plt.legend(loc=4)
plt.savefig('sample.jpg',dpi=350)
plt.grid()
plt.show()
```

```python
# instead of lines i want marker
plt.plot(score,label='Maharashtra',color='#761377',marker='o')# dashed lines
plt.plot(score2,label='Andhra',color='#146f64',marker='o')
plt.xticks(ticks=range(6),labels=['Day1','Day2','Day3','Day4','Day5','Day6'])
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Day vs Temp')
plt.legend(loc=4)
plt.savefig('sample.jpg',dpi=350)
plt.grid()
plt.show()
```

```python
plt.plot(score,label='Maharashtra',color='#761377',marker='o',linestyle=' ')# dashed lines
plt.plot(score2,label='Andhra',color='#146f64',marker='o',linestyle=' ')
plt.xticks(ticks=range(6),labels=['Day1','Day2','Day3','Day4','Day5','Day6'])
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Day vs Temp')
plt.legend(loc=4)
plt.savefig('sample.jpg',dpi=350)
plt.grid()
plt.show()
```

## Bivariate line plot

```python
# power play over
over = [1,2,3,4,5]#X:independant
runs = [4,10,8,18,20]#y: dependant
```

```python
import pandas as pd
p = pd.DataFrame({'over':over,'runs':runs})
p
```

```python
plt.plot(over,runs)
```

```python
import numpy as np
np.cumsum(runs)
```

```python
plt.plot(over,np.cumsum(runs))
plt.xlim((1,10))
```
